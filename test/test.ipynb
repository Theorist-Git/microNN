```python
import sys
import os
import importlib

sys.path.append(os.path.abspath('..'))

import MLP.nn as nn
importlib.reload(nn)

MLP = nn.MLP
```

```python
# Regression
# Inputs: features (x1, x2)
xs = [
  [1.0, 2.0],
  [2.0, 3.0],
  [3.0, 4.0],
  [4.0, 5.0],
]

# Targets: continuous values (e.g., y = 2*x1 + 3*x2)
ys = [8.0, 13.0, 18.0, 23.0]
```

```python
import numpy as np

xs = np.array(xs)  # Convert list of lists to numpy array
xs = (xs - xs.mean(axis=0)) / xs.std(axis=0)
xs = xs.tolist()
```

```python
reg_nn = MLP(
    n_inputs=2, 
    layers=[
        (2, "relu"),
        (1, "linear"),
    ]
)
```

```python
reg_nn.parameters()
```

```python
reg_nn.layers[0].activation
```

```python
EPOCHS = 1000
lr     = 0.001 
```

```python
for k in range(EPOCHS):

    # forward propagation
    y_pred = [reg_nn(x) for x in xs]
    loss = sum([(y_hat - y) ** 2 for y, y_hat in zip(ys, y_pred)]) / len(ys)

    # backpropagation
    reg_nn.zero_grad()
    loss.backward()

    # update
    for p in reg_nn.parameters():
        p.data -= lr * p.grad

    print(f"EPOCH {k}: Mean Squared Loss = {loss.data}")
```

```python
y_pred = [reg_nn(x) for x in xs]
y_pred
```
